```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from scipy.stats import truncnorm, norm
from sklearn.model_selection import train_test_split

import time
```

```python
data = pd.read_table("german.data-numeric", header=None, delim_whitespace=True)
```

```python
# separating predictors from target and transforming the target labels from 1-2 to 0-1
X, y = data[data.columns[:-1]], data[data.columns[-1]] - 1
# normalizing the predictors
X = (X - X.mean(axis=0))/X.std(axis=0)
# adding the bias column
X.insert(0, "bias", 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
```

```python
def truncated_normal(mean, y):
    """Used to sample z given beta
    needs a :
    - mean 
    - support borders where to sample
    """
    a = -mean
    b = -mean
    
    a[y<=0] = -np.inf
    b[y>0] = np.inf
    
    z = truncnorm.rvs(a, b, loc=mean, size=(len(mean),))
    
    return z

def Gibbs_Sampling(data, target, iterations, tau=100, burnin=3000):
    """Performs Gibbs sampling iteratively from z given beta,X,Y and beta given z,X,Y 
    to approximate samples from the join (z, beta) given (X,Y)
    returns : beta_burnin, beta_after_burnin, z_burnin, z_after_burnin
              Those are the samples in each direction (variable)
    """
    
    n = data.shape[0]
    dim = data.shape[1]
    sigma = np.linalg.inv((np.eye(dim) / tau ) + data.T @ data)
    used_to_speed_up_things = sigma @ data.T
    
    z = np.random.randn(n) # initialisation of z0 : a vector of size the training data sampled from the normal distribution
    list_z = []
    list_beta = []
    
    
    for t in range(burnin + iterations):
        mu = used_to_speed_up_things @ z
        beta = np.random.multivariate_normal(mu, cov=sigma) # beta given z sampling
        list_beta.append(beta)
        z = truncated_normal( data @ beta, target)
        list_z.append(z)
    
    return np.array(list_beta)[:burnin], np.array(list_beta)[burnin:], np.array(list_z)[:burnin], np.array(list_z)[burnin:]

def prediction_phase(data_test, betas_post_burnin):
    
    z_pred = data_test @ betas_post_burnin.T
    noise = np.random.randn(z_pred.shape[0], z_pred.shape[1])    
    return np.sign(z_pred+noise).replace(-1,0)
```

```python
iterations = 5000
burnin = 3000
start = time.time()
betas_burnin, betas_post_burnin, z_burnin, z_post_burnin = Gibbs_Sampling(data = X_train,
                                                                          target=y_train,
                                                                          iterations=iterations,
                                                                          burnin=burnin)
end = time.time()
print("Speed : ", end - start )
```

```python
plt.figure(figsize=(15, 8))
index_to_show = np.random.randint(low=0, high= X_train.shape[1]-1)

plt.subplot(2,1,1)
plt.title(r"$\beta_{"+str(index_to_show)+"}$")
plt.plot(np.arange(burnin), 
         betas_burnin[:,index_to_show], color = "r", label = "burnin");
plt.plot(np.arange(burnin, iterations+burnin), 
         betas_post_burnin[:,index_to_show], label = "after burnin");
plt.legend()

plt.subplot(2,1,2)
plt.title(r"$z_{"+str(index_to_show)+"}$")
plt.plot(np.arange(burnin), 
         z_burnin[:,index_to_show], color = "r", label = "burnin");
plt.plot(np.arange(burnin, iterations+burnin), 
         z_post_burnin[:,index_to_show], label = "after burnin");
plt.legend()

plt.figure(figsize=(15, 8))
plt.subplot(1,2,1)
plt.title(r"$\beta_{"+str(index_to_show)+"}$ distribution after burnin phase")
plt.hist(betas_post_burnin[:,index_to_show], bins=20, edgecolor = "k");
plt.subplot(1,2,2)
plt.title(r"$z_{"+str(index_to_show)+"}$ distribution after burnin phase")
plt.hist(z_post_burnin[:,index_to_show], bins=20, edgecolor = "k");
```

```python
y_pred = prediction_phase(X_test, betas_post_burnin)
y_pred = (y_pred.mean(axis=1) > 0.5)*1
```

```python
(y_pred == y_test).mean()
```

# Mean Field variational sampling

We use the fact that the mean of a truncated gaussian is :

$$ X \beta + y \frac{\phi(X \beta)}{\Phi(yX\beta)}$$ 

where $\phi$ is the pdf of the normal distribution and $\Phi$ is the CDF

```python
def mean_field(data, target, iterations, tau=100):
    
    n = data.shape[0]
    d = data.shape[1]
    sigma = np.linalg.inv(data.T @ data + (np.eye(d) / tau))
    
    betas = np.zeros((iterations, d))
    zs = np.zeros((iterations, n))
    beta = np.random.randn(d)[:,None] #initialization of beta
    
    for t in range(iterations):
        betas[t] = beta.ravel()
        mean = data @ beta.ravel()
        z =  mean + target * norm.pdf(mean) / norm.cdf(target * mean) #update z
        zs[t] = z
        beta = sigma @ (data.T @ z)#update beta
        
    return betas, zs
```

```python
start = time.time()
betas_means, zs_means = mean_field(data = X_train,
                               target= y_train,
                               iterations=500)
end = time.time()
print("Speed : ", end - start )
fig = plt.figure(figsize=(14,6))
plt.suptitle("the L2 norm of differences of the successive means computed")
plt.subplot(1,2,1)
plt.title(r"$\beta$")
plt.grid()
plt.plot(np.linalg.norm(np.diff(betas_means, axis = 0), axis = 1))
plt.subplot(1,2,2)
plt.title(r"z")
plt.grid()
plt.plot(np.linalg.norm(np.diff(zs_means, axis = 0), axis = 1));
```

```python
def sampling(mean_beta, mean_z, data, target, nb_samples, tau=100):
    
    n = data.shape[0]
    d = data.shape[1]
    sigma = np.linalg.inv(data.T @ data + np.eye(d) / tau)
    betas = np.random.multivariate_normal(mean=mean_beta,
                                          cov=sigma,
                                          size=nb_samples)
    z_s = np.zeros((nb_samples, n))
    for t in range(nb_samples):
        z_s[t] = truncated_normal(mean_z, target)
    return betas, z_s
```

```python
betas, z_s = sampling(mean_beta=betas_means[-1],
                     mean_z = zs_means[-1],
                     data= X_train,
                     target= y_train,
                     nb_samples=5000)
```

```python
i = np.random.choice(range(25))
fig = plt.figure(figsize=(11,6))
plt.title(r" sampled $\beta_{"+str(i)+"}$")
plt.grid()
plt.hist(betas[:,i], edgecolor = "k", bins=20);
```

```python
def prediction(beta_sampled, X_test):
    n = X_test.shape[0]
    d = X_test.shape[1]
    z_ = X_test @ beta_sampled.T
    y_ = np.sign(z_.T)
    return y_
```

```python
y_pred = prediction(beta_sampled=betas, X_test=X_test)
y_pred = (y_pred.mean(axis=0) < 0.5)*1
```
